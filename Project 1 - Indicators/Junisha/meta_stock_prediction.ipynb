```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('/content/META Historical Data (1).xlsx')
# Assuming the data is in the first column and tab-separated, split it into multiple columns
df = df[df.columns[0]].str.split('\t', expand=True)
# Rename the columns based on the first row
df.columns = df.iloc[0]
# Drop the first row as it is now the header
df = df[1:]
df
```

```python
df = df.drop(df.index[0:4])
```

```python
df.describe()
```

```python
# Clean and convert price columns to numeric
df['Open'] = df['Open'].replace({r'\$': ''}, regex=True).astype(float)
df['Close/Last'] = df['Close'].replace({r'\$': ''}, regex=True).astype(float) # Corrected column name to 'Close'

# Convert 'Date' to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Sort by date
df = df.sort_values(by='Date')

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(df['Date'], df['Open'], label='Open Price')
plt.plot(df['Date'], df['Close/Last'], label='Close Price')

plt.title('Open vs Close Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ensure the DataFrame has the correct column names and is in the correct order
df = df[df.columns[0]].str.split('\t', expand=True)
df.columns = df.iloc[0]
df = df[1:].copy()
df = df.iloc[::-1].reset_index(drop=True) # Reverse the order and reset index

# Clean and convert 'High' column to numeric
df['High'] = df['High'].replace({r'\$': ''}, regex=True).astype(float)

# Convert 'Date' to datetime objects and extract month and year
df['Date'] = pd.to_datetime(df['Date'])
df['Month_Year'] = df['Date'].dt.to_period('M')

# Find the highest price for each month
monthly_highs = df.groupby('Month_Year')['High'].max().reset_index()

# Convert Month_Year back to string for plotting
monthly_highs['Month_Year'] = monthly_highs['Month_Year'].astype(str)

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Month_Year', y='High', data=monthly_highs)

plt.title('Highest Price per Month')
plt.xlabel('Month and Year')
plt.ylabel('Highest Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Extract month and year
df['Month_Year'] = df['Date'].dt.to_period('M')

# Calculate total volume per month
monthly_volume = df.groupby('Month_Year')['Volume'].sum().reset_index()

# Convert Month_Year back to string for plotting
monthly_volume['Month_Year'] = monthly_volume['Month_Year'].astype(str)

# Create the horizontal bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Volume', y='Month_Year', data=monthly_volume, orient='h')

plt.title('Total Volume Sold per Month')
plt.xlabel('Total Volume')
plt.ylabel('Month and Year')
plt.tight_layout()
plt.show()
```
